```python
# Import Libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# load dataset 
df = pd.read_pickle(r"C:\Users\Windows\Instacart Basket Analysis\02 Data\Prepared Data\df_small_updated.pkl")
```

```python
path = r"C:\Users\Windows\Instacart Basket Analysis"
```

```python
# distribution of income groups by Region 

ct = pd.crosstab(df['Region'], df['income_group'], normalize = True)
```

```python
ct
```

```python
fig_6 = sns.heatmap(ct, annot=True, linewidths=0.5, cmap='crest' )
```

```python
fig_6.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'fig6_income_region.png'))
```

```python
# The customers are in the high income and middle income groups in the South, followed by the West, Midwest, and Northeast
```

```python
# aggregating by days since prior order
df.groupby(['income_group', 'Region'])['days_since_prior_order'].agg(['max','min','mean'])
```

```python
# distribution of customers by income for average # of days since prior order across 4 regions

fig_8 = sns.barplot(data = df, x='income_group', y='days_since_prior_order', hue='Region', palette='colorblind')
```

```python
# save figure 
fig_8.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'fig8_days_since_prior_income_region.png'))
```

```python
# aggregating by expenditure based on income group and region
df.groupby(['income_group', 'Region'])['prices'].agg(['max','min','mean'])
```

```python
# distribution of customers by income for average expenditures across 4 regions

fig_x = sns.barplot(data = df, x='income_group', y='prices', hue='Region', palette='colorblind')
```

```python
fig_x.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'fig7_expenditure_income_region.png'))
```

### Step 6

```python
# find the mean of prices paid by each user_id 

ords_prods_merge['avg_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(5)
```

```python
# create flag for low spender
ords_prods_merge.loc[ords_prods_merge['avg_spending'] < 10, 'spender_flag'] = 'Low spender'
```

```python
# create flag for high spender
ords_prods_merge.loc[ords_prods_merge['avg_spending'] >= 10, 'spender_flag'] = 'High spender'
```

```python
# checking for new columnns created: spender_flag and average price paid by each user_id
ords_prods_merge.head(50)
```

### Step 7

```python
# adding a new column for the median number of days since last order for each user_id

ords_prods_merge['last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# checking  addition of the column for the median of days since prior order

ords_prods_merge.info()
```

```python
#  create flag for Non-frequent customer
ords_prods_merge.loc[ords_prods_merge['last_order'] > 20, 'ord_freq_flag'] = 'Non-frequent customer'
```

```python
#  create flag for regular customer
ords_prods_merge.loc[(ords_prods_merge['last_order'] > 10) & (ords_prods_merge['last_order'] <=20), 'ord_freq_flag'] = 'regular customer'
```

```python
# create flag for frequent customer
ords_prods_merge.loc[ords_prods_merge['last_order'] <= 10, 'ord_freq_flag'] = 'Frequent customer'
```

```python
# checking the addition of the flag column.
ords_prods_merge.head()
```

```python
# checking maximum days since last order
ords_prods_merge['last_order'].max()
```

```python
# checking accuracy of the ord_freq flag. It works. Users with more than 20 days since last order are flagged as non-frequent customers 

ords_prods_merge[ords_prods_merge['ord_freq_flag'] =='Non-frequent customer'].head(10)
```

```python
# Checking the accuracy of the spender flag. Yes, it looks like "high spenders" are associated with purchases of more than $10.

ords_prods_merge[ords_prods_merge['spender_flag'] == 'High spender'].head(10)
```

```python
# Checking the accuracy of the spender flag. Yes it works. For avg spending > 10, the spender flag is high spender. 
ords_prods_merge[ords_prods_merge['avg_spending'] > 10 ].head(10)
```
